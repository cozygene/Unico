```r
pval.list  = list()
coeff.list = list()

for (data.name in data.names){

    tcax.mdl      = readRDS(file.path(results_dir, data.name,'tcax.mdl.mp.0.vp.0.cp.0.tau.FALSE.max_stds.2.rds'))
    tca.mdl       = readRDS(file.path(results_dir, data.name,'tca.mdl.rds'))
    celldmc.mdl  = readRDS(file.path(results_dir,  data.name,'CellDMC.mdl.rds'))
    #baseline.mdl = readRDS(file.path(results_dir, "liu",'Baseline.rds'))
    
    pval  = list()
    coeff = list()
    
    #TCAx
    for (slot.name in c("parametric.full", "parametricX2Q2", "parametricXQ2")){
        
        pval [[paste0("TCAx.",slot.name,".marginal")]] = tcax.mdl$params.hat[[slot.name]]$gammas_hat_pvals[,paste(source.ids, pheno, sep=".")]
        coeff[[paste0("TCAx.",slot.name,".marginal")]] = tcax.mdl$params.hat[[slot.name]]$gammas_hat[,paste(source.ids, pheno, sep=".")]
        
        pval [[paste0("TCAx.",slot.name,".joint")]] = tcax.mdl$params.hat[[slot.name]]$gammas_hat_pvals.joint[, pheno, drop = F]
        coeff[[paste0("TCAx.",slot.name,".joint")]] = (pval [[paste0("TCAx.",slot.name,".joint")]] > -1) * 1
    }
    
    #TCA
    for (slot.name in c("parametric.full", "parametricX2", "parametricXQ2")){
        
        pval [[paste0("TCA.",slot.name,".marginal")]] = tca.mdl$params.hat[[slot.name]]$gammas_hat_pvals[,paste(source.ids, pheno, sep=".")]
        coeff[[paste0("TCA.",slot.name,".marginal")]] = tca.mdl$params.hat[[slot.name]]$gammas_hat[,paste(source.ids, pheno, sep=".")]
        
        pval [[paste0("TCA.",slot.name,".joint")]] = tca.mdl$params.hat[[slot.name]]$gammas_hat_pvals.joint[, pheno, drop = F]
        coeff[[paste0("TCA.",slot.name,".joint")]] = (pval [[paste0("TCA.",slot.name,".joint")]] > -1) * 1
    } 
    
    #CellDMC
    for (slot.name in c("parametric.full", "parametricX2")){
        slot.res = celldmc.mdl[[pheno]][[slot.name]]
        pval [[paste0("CellDMC.",slot.name,".marginal")]] = cbind(slot.res$marg.res$Gran[,"p"], 
                                                                  slot.res$marg.res$CD4T[,"p"],
                                                                  slot.res$marg.res$CD8T[,"p"],
                                                                  slot.res$marg.res$Mono[,"p"],
                                                                  slot.res$marg.res$NK[,"p"],
                                                                  slot.res$marg.res$B[,"p"])
        colnames(pval [[paste0("CellDMC.",slot.name,".marginal")]]) = source.ids
        
        coeff[[paste0("CellDMC.",slot.name,".marginal")]] = cbind(slot.res$marg.res$Gran[,"Estimate"], 
                                                                  slot.res$marg.res$CD4T[,"Estimate"],
                                                                  slot.res$marg.res$CD8T[,"Estimate"],
                                                                  slot.res$marg.res$Mono[,"Estimate"],
                                                                  slot.res$marg.res$NK[,"Estimate"],
                                                                  slot.res$marg.res$B[,"Estimate"])
        colnames(coeff [[paste0("CellDMC.",slot.name,".marginal")]]) = source.ids
        
        
        pval [[paste0("CellDMC.",slot.name,".joint")]] = slot.res$joint.res
        colnames(pval [[paste0("CellDMC.",slot.name,".joint")]]) = c(pheno)
        coeff[[paste0("CellDMC.",slot.name,".joint")]] = (pval [[paste0("CellDMC.",slot.name,".joint")]] > -1) * 1
    } 
    
    pval.list[[data.name]]  = pval
    coeff.list[[data.name]] = coeff
}
```

```r
calc_meta_pvals = function(pvals){
    #https://en.wikipedia.org/wiki/Fisher%27s_method
    np = length(pvals)
    chi.stats = - 2 * (sum(log(pvals)))
    return(pchisq(q=chi.stats , df=2 * np, lower.tail=F))
}
```

```r
#pvals_mats: a list of p value matrices. each matrix is from one experiment
#returns a meta_pvals_mat: a single matrix on shared feature.ids and source.ids, 
# each entry is the meta p val calculated across all experiments
calc_meta_pvals_mat = function(pvals_mats){
    #shared feature.ids and source.ids
    feature.ids = c()
    source.ids = c()
    for (t in 1:length(pvals_mats)){
        if (t == 1){
            feature.ids = rownames(pvals_mats[[t]])
            source.ids = colnames(pvals_mats[[t]])
        }else{
            feature.ids = intersect(feature.ids, rownames(pvals_mats[[t]]))
            source.ids  = intersect(source.ids,  colnames(pvals_mats[[t]]))
        }
    }
    
    pvals_array = array(1, c(length(pvals_mats), length(feature.ids), length(source.ids)))
    dimnames(pvals_array)[[1]] = paste0("experiment", 1:length(pvals_mats))
    dimnames(pvals_array)[[2]] = feature.ids
    dimnames(pvals_array)[[3]] = source.ids
    for (t in 1:length(pvals_mats)){
        pvals_array[t,,] = pvals_mats[[t]][feature.ids, source.ids]
    }

    meta_pvals_mat = matrix(1, length(feature.ids), length(source.ids))
    rownames(meta_pvals_mat) = feature.ids
    colnames(meta_pvals_mat) = source.ids

    for (feature.id in 1:length(feature.ids)){
        for (source.id in 1:length(source.ids)){
            meta_pvals_mat[feature.id,source.id] = calc_meta_pvals(pvals = pvals_array[,feature.id,source.id])
        }
    }

    return(meta_pvals_mat)
}
```

```r
meta_pvals_mat = calc_meta_pvals_mat(pvals_mats)
```

# Testing 

```r
get_confusion_df = function(dis_pvals_mat, val_pvals_mat,
                            dis_coeffs_mat, val_coeffs_mat,
                            sig_thr,  dis_sig_mode,  val_sig_mode,
                            null_thr, dis_null_mode, val_null_mode, verbose = F){
    # "strict": bonforroni
    # "loose": number of cpgs in discovery
    # "unadjust": no adjustment 
    # always want the sign signs to agree as well to be call TP

    #first make sure the columns are the same
    #keep only the shared cpgs
    
    
    
    #automatically agrees on sign
    #dis_coeffs_mat [,] = 1
    #val_coeffs_mat [,] = 1
    
    message(paste("discovery shape: ", dim(dis_pvals_mat)[1], dim(dis_pvals_mat)[2]))
    message(paste("validation shape: ", dim(val_pvals_mat)[1], dim(val_pvals_mat)[2])) 
     
    keep.probs     = intersect(rownames(dis_pvals_mat), rownames(val_pvals_mat))
    dis_pvals_mat  = duplicate(dis_pvals_mat)[keep.probs, , drop = F]
    dis_coeffs_mat = duplicate(dis_coeffs_mat)[keep.probs, , drop = F]
    val_pvals_mat  = duplicate(val_pvals_mat)[keep.probs, , drop = F]
    val_coeffs_mat = duplicate(val_coeffs_mat)[keep.probs, , drop = F]
    
    message("after")
    message(paste("discovery shape: ", dim(dis_pvals_mat)[1], dim(dis_pvals_mat)[2]))
    message(paste("validation shape: ", dim(val_pvals_mat)[1], dim(val_pvals_mat)[2])) 
    message("###################")
    
    #Positive
    if (dis_sig_mode == "strict"){
        dis_sig_thr = sig_thr/(dim(dis_pvals_mat)[1] * dim(dis_pvals_mat)[2])
    }else if (dis_sig_mode == "unadjust"){
        dis_sig_thr = sig_thr
    }else{
        print("error")
    }
    if(verbose){print(paste0("dis_sig_thr: ",  dis_sig_thr))}

    dis_sig_mask = dis_pvals_mat < dis_sig_thr
    n_ds = sum(dis_sig_mask)
    if(verbose){print(paste0("n_ds: ",  n_ds))}

    if (val_sig_mode == "strict"){
        val_sig_thr = sig_thr/(dim(val_pvals_mat)[1] * dim(val_pvals_mat)[2])
    }else if (val_sig_mode == "loose"){
        val_sig_thr = sig_thr/n_ds
    }else if (val_sig_mode == "unadjust"){
        val_sig_thr = sig_thr
    }else{
        print("error")
    }
    if(verbose){print(paste0("val_sig_thr: ",  val_sig_thr))}
    
    val_sig_mask_pvals = val_pvals_mat[dis_sig_mask] < val_sig_thr
    print("val_sig_mask_pvals")
    print(sum(val_sig_mask_pvals))
    
    val_sig_mask_ceoffs = sign(val_coeffs_mat[dis_sig_mask]) == sign(dis_coeffs_mat[dis_sig_mask]) #(same sign)
    print("val_sig_mask_ceoffs")
    print(sum(val_sig_mask_ceoffs))
    val_sig_mask = val_sig_mask_pvals & val_sig_mask_ceoffs
    TP = sum(val_sig_mask)
    FN = sum(!val_sig_mask)


    # Negative
    if(dis_null_mode == "strict"){
        dis_null_thr = null_thr/(dim(dis_pvals_mat)[1] * dim(dis_pvals_mat)[2])
    }else if (dis_null_mode == "unadjust"){
        dis_null_thr = null_thr
    }else{
        print("error")
    }
    if(verbose){print(paste0("dis_null_thr: ",  dis_null_thr))}

    dis_null_mask = dis_pvals_mat > dis_null_thr

    n_dn = sum(dis_null_mask)
    if(verbose){print(paste0("n_dn: ",  n_dn))}

    if (val_null_mode == "strict"){
        val_null_thr = null_thr/(dim(val_pvals_mat)[1] * dim(val_pvals_mat)[2])
    }else if (val_null_mode == "loose"){
        val_null_thr = null_thr/n_dn
    }else if (val_null_mode == "unadjust"){
        val_null_thr = null_thr
    }else{
        print("error")
    }
    if(verbose){print(paste0("val_null_thr: ",  val_null_thr))}

    val_null_mask = val_pvals_mat[dis_null_mask] > val_null_thr 

    FP = sum(!val_null_mask)
    TN = sum(val_null_mask)   
    
    recall    = TP/(TP + FN)
    precision = TP/(TP + FP)
    specificity = TN/(TN + FP)
    F1        = 1/(0.5 * (1/recall + 1/precision))
    accuracy = (TP + TN)/(TP + FP + TN + FN)
    
    TP = as.numeric(TP)
    TN = as.numeric(TN)
    FP = as.numeric(FP)
    FN = as.numeric(FN)
    MCC = as.numeric((TP*TN) - (FP*FN))/ sqrt((TP+FP)*(TP+FN)*(TN+FP)*(TN+FN))
    #print(paste(TP, TN, FP,FN ))
    confusion_df = data.frame(TP = TP, 
                              FN = FN, 
                              FP = FP, 
                              TN = TN, 
                              recall = recall, 
                              precision = precision, 
                              F1 = F1,
                              specificity = specificity,
                              accuracy = accuracy,
                              MCC = MCC)
    return(confusion_df)
}
```

```r
methods = c()
for (data.name in names(pval.list)){
    methods = c(methods, list(names(pval.list[[data.name]])))
}
methods = Reduce(intersect, methods)
print("methods with result on all 4 datasets")
print(methods)
```

# get this across all dataset, across all methods

```r
confusion_dfs_list = list()
for (val.data.name in data.names){
    confusion_dfs = c()
    for (method in methods){
        val_pvals_mat = pval.list[[val.data.name]][[method]]

        dis_pvals_mats = list()
        for (dis.data.name in data.names){
            if (dis.data.name != val.data.name){
                dis_pvals_mats[[dis.data.name]] = pval.list[[dis.data.name]][[method]]
            }
        }
        meta_pvals_mat = calc_meta_pvals_mat(dis_pvals_mats)

        confusion_df = get_confusion_df(meta_pvals_mat, val_pvals_mat,
                                        meta_pvals_mat, val_pvals_mat, # essentially turn off this check 
                                        sig_thr = 0.05,  dis_sig_mode = "strict",   val_sig_mode = "loose",
                                        null_thr= 0.95,  dis_null_mode= "unadjust", val_null_mode = "loose", verbose = verbose)
        rownames(confusion_df) = c(method)
        confusion_dfs = rbind(confusion_dfs, confusion_df)
    }
    confusion_dfs_list[[val.data.name]] = confusion_dfs
}

# make it an array 
confusion_dfs_array = array(0,c(length(data.names), length(methods), ncol(confusion_dfs_list[[1]])))
dimnames(confusion_dfs_array)[[1]] = data.names
dimnames(confusion_dfs_array)[[2]] = methods
dimnames(confusion_dfs_array)[[3]] = colnames(confusion_dfs_list[[1]])

for(data.name in data.names){
    confusion_dfs_array[data.name,,]  =  as.matrix(confusion_dfs_list[[data.name]])
}
```
