# baseline testing

```r
#x: a vector of numeric number
#std.thr: numeric
#min.thr: numeric, absolute minimal thr, if not provided, by default set to minimal number in "x"
#max.thr: numeric, absolute maximal thr, if not provided, by default set to maximal number in "x"
#returns a vector of logical, same length of input "x". 
#only 1 if that number is within std.thr standard deviation and within absolute range min.thr and max.thr
#otherwise set to 0 
mask_outliers = function(x, std.thr, min.thr = NULL, max.thr = NULL){
    if(is.null(min.thr)){min.thr = min(x)}
    if(is.null(max.thr)){max.thr = max(x)}
    mu = mean(x)
    sigmas = sd(x)
    min.std.thr = mu - std.thr * sigmas 
    max.std.thr = mu + std.thr * sigmas 

    mask = (x <= max.thr) &  (x >= min.thr) & (x <= max.std.thr) & (x >= min.std.thr) 
    return(mask)
}
```

```r
#X: matrix of features by samples, methylation beta values
#W: matrix of number of samples by number of celltypes. 
#C: matrix of number of samples by number of covariates, both global and celltype level covars. these will be the phenotypes as well
#max_stds: numberic, samples fall out of the max_stds will be discarded
#this functions essentially fit Z.hat(XW) ~ 1 + W + C for marginal test
#this functions essentially fit X ~ 1 + W + C for joint test

#returns a list of 6 keys
#first three keys:  marg.pvals (k by m by number of covars), marg.coefsf (k by m by number of covars), marg.masks(k by m by n),
#second three keys: joint.pvals(m by number of covars), joint.coefsf(m by number of covars). joint.masks(m by n),

base_assos = function(X, W, C, max_stds){

    source.ids  = colnames(W)
    feature.ids = rownames(X)
    sample.ids  = colnames(X)
    covar.ids   = colnames(C)


    # marginal related
    marg.pvals = array(1, c(length(source.ids), length(feature.ids), length(covar.ids)))
    dimnames(marg.pvals)[[1]] = source.ids
    dimnames(marg.pvals)[[2]] = feature.ids
    dimnames(marg.pvals)[[3]] = covar.ids
    marg.coefs = copy(marg.pvals)

    marg.masks = array(0, c(length(source.ids), length(feature.ids), length(sample.ids)))
    dimnames(marg.masks)[[1]] = source.ids
    dimnames(marg.masks)[[2]] = feature.ids
    dimnames(marg.masks)[[3]] = sample.ids 


    # joint related
    joint.pvals = matrix(1, length(feature.ids), length(covar.ids))
    rownames(joint.pvals) = feature.ids
    colnames(joint.pvals) = covar.ids
    joint.coefs = copy(joint.pvals)

    joint.masks = matrix(0, length(feature.ids), length(sample.ids))
    rownames(joint.masks) = feature.ids
    colnames(joint.masks) = sample.ids


    for (feature.id in feature.ids){    
        # marginal model
        for (source.id in source.ids){
            Z.hat = X[feature.id, sample.ids] * W[sample.ids, source.id]
            df   = data.frame(y = Z.hat, 
                              cbind(W[sample.ids,-1], C[sample.ids,])) # drop W to prevent co-linear
            masks = mask_outliers(df$y,  std.thr = max_stds)
            marg.masks[source.id, feature.id, sample.ids] = masks
            df   = df[masks, ]

            mdl.fit = lm(y ~ 1 + ., data = df) 
            marg.pvals[source.id, feature.id, covar.ids] = summary(mdl.fit)$coef[covar.ids, "Pr(>|t|)"]
            marg.coefs[source.id, feature.id, covar.ids] = summary(mdl.fit)$coef[covar.ids, "Estimate"]   
        }

        df   = data.frame(y = as.vector(X[feature.id, sample.ids, drop = F]), 
                          cbind(W[sample.ids, -1], C[sample.ids,])) # drop W to prevent co-linear
        masks = mask_outliers(df$y,  std.thr = max_stds)
        joint.masks[feature.id, sample.ids] = masks
        df   = df[masks, ]

        # joint model
        mdl.fit = lm(y ~ 1 + ., data = df) 
        joint.pvals[feature.id, covar.ids] = summary(mdl.fit)$coef[covar.ids, "Pr(>|t|)"]
        joint.coefs[feature.id, covar.ids] = summary(mdl.fit)$coef[covar.ids, "Estimate"] 


    }


    res = list(marg.pvals = marg.pvals, marg.coefs = marg.coefs, marg.masks = marg.masks,
               joint.pvals = joint.pvals, joint.coefs = joint.coefs, joint.masks = joint.masks)
    
    return(res)
}
    
```

# confirms the masking is on the X

```r
parametric.full = base_assos(X, W, C, max_stds = Inf)
parametric.X2   = base_assos(X, W, C, max_stds = 2)
```

```r
j = 10
X.j.mean = mean(X[j, ])
X.j.sd = sd(X[j, ])
masks = parametric.X2$joint.masks
```

```r
# all masked sample in at feature j are 2 std out
all((X[j,masks[j,] == 0] <= (X.j.mean - 2*X.j.sd)) | (X[j,masks[j,] == 0] >= (X.j.mean + 2*X.j.sd)))
```

```r
# all preserved sample in at feature j are 2 std in
all((X[j,masks[j,] == 1] >= (X.j.mean - 2*X.j.sd)) & (X[j,masks[j,] == 1] <= (X.j.mean + 2*X.j.sd)))
```

```r
j = 10
h = 3
# that source's Z
Z = X *repmat(t(W[,h,drop = F]), nrow(X), 1)
Z.j.mean = mean(Z[j, ])
Z.j.sd = sd(Z[j, ])

masks = parametric.X2$marg.masks[h,,]
```

```r
all((Z[j,masks[j,] == 1] >= (Z.j.mean - 2*X.j.sd)) & (Z[j,masks[j,] == 1] <= (Z.j.mean + 2*Z.j.sd)))
```

```r
all((Z[j,masks[j,] == 0] <= (Z.j.mean - 2*X.j.sd)) | (Z[j,masks[j,] == 0] >= (Z.j.mean + 2*Z.j.sd)))
```
