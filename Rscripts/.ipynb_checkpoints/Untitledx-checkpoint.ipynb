```r
library("matrixStats")# colMeans

library("ggplot2")
library("ggpubr")

source("analysis.utils.r")
source("simulate.expression.utils.r")

set.seed(2023)
```

```r
source.col = "decon.L1" 
#source.col = "decon.L2" 

max_stds = 2
#max_sds = 3

N = 500

if (source.col == "decon.L1"){
    k = 5
}else{
    k = 7
}

cl = (k*k - k)/2 #covar lenght
```

```r
project.dir = "/u/home/j/johnsonc/project-halperin/TCAx/TCAx2023/"

#relative path of the data dir to project.dir
if(source.col == "decon.L1"){
    data.dir = paste0("Data/RNA/Simulation-PBMC/sc-Stephenson_all.W_decon.L1_HEF.10k_k_5_m_600_n_", N, "_dirichlet_F_noiseZ_T_varThr_1e-04_filThr_1e-04_expQtl_0_enrich_F_etpRat_0_enrichRat_0_maxSds_", max_stds, "_scale.maxSds_Inf_scale.factor.thr_1e-04/")
    figure.dir  = paste0("Figure/Simulation-PBMC/decon.L1.max_stds.",max_stds)
}else{
    data.dir = paste0("Data/RNA/Simulation-PBMC/sc-Stephenson_all.W_decon.L2_HEF.10k_k_7_m_600_n_", N, "_dirichlet_F_noiseZ_T_varThr_1e-04_filThr_1e-04_expQtl_0_enrich_F_etpRat_0_enrichRat_0_maxSds_", max_stds, "_scale.maxSds_Inf_scale.factor.thr_1e-04/")
    figure.dir  = paste0("Figure/Simulation-PBMC/decon.L2.max_stds.",max_stds)
}

ts = 1:20
```

```r
data.dir   = file.path(project.dir, data.dir)
figure.dir = file.path(project.dir, figure.dir)
data.name  = strsplit(data.dir, "/")[[1]]
res.dir    = file.path(project.dir, paste(c("Result", data.name[(length(data.name)-2) : length(data.name)]), collapse = "/"))

if (!file.exists(res.dir)){print("no result in the result directory")}
if (!file.exists(figure.dir)) {dir.create(figure.dir, recursive = T)}
print(data.dir)
print(res.dir)
print(figure.dir)
```

# load figures

```r
fig.list.file   = file.path(figure.dir, paste0("RNA_Simulation-PBMC_", source.col,"_", N, "_fig.list.rds"))
stats.list.file = file.path(figure.dir, paste0("RNA_Simulation-PBMC_", source.col,"_", N, "_stats.list.rds"))

if (file.exists(fig.list.file)){
    fig.list = readRDS(fig.list.file)
}else{
    print("no fig.list.file")
}
  
if (file.exists(stats.list.file)){
    stats.list = readRDS(stats.list.file)
}else{
    print("no stats.file")
}   
```

```r
names(fig.list)
```

```r
plts = list(3)
plts[[1]] = fig.list$mean.bar.g + theme(legend.position="none") 
plts[[2]] = fig.list$var.bar.g + theme(legend.position="none") + theme(axis.title.y = element_blank()) 
plts[[3]] = fig.list$covar.bar.g + theme(axis.title.y = element_blank()) 

covar.width  = round(2 * cl/k)
params.bar.g = egg::ggarrange(plots=list(plts[[1]], plts[[2]], plts[[3]]),  
                   labels = c("a", " b", "c"), 
                   align = "h",
                   widths = c(2,2,covar.width),  
                   label.args = list(gp = grid::gpar(font = 20, cex =3)), 
                   debug=F)
```

```r
cor.box.g = egg::ggarrange(plots = fig.list$Z.corrs.g, 
                   labels = letters[4: (4+k-1)], 
                   align = "h",
                   widths = rep(2, ceil(k/2)),  
                   label.args = list(gp = grid::gpar(font = 20, cex =3)), 
                   debug=F)
```

```r
add.box.g = egg::ggarrange(plots = list(fig.list$add.box.g), 
                           labels = letters[k + 4], 
                           label.args = list(gp = grid::gpar(font = 20, cex =3)), 
                           debug=F)
```

```r
options(repr.plot.width = 24, repr.plot.height = 30, repr.plot.res = 200)
multi.g = ggarrange(params.bar.g,
                    cor.box.g,
                    add.box.g, 
                    nrow = 3,
                    heights = c(1,2.25,1),
                    #labels = c("", "", letters[k + 3]),
                    font.label = list(size = 25, color = "black", face = "bold", family = NULL))
```

```r
ggsave(file.path(figure.dir, paste0("RNA_Simulation-PBMC_", source.col,"_", N, "_multi.pdf")), multi.g,
       device = "pdf", width = 24, height = 30, dpi = 600)
```
