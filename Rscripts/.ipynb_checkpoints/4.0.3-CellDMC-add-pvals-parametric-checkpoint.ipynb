# reimplement

```r
#x: a vector of numeric number
#std.thr: numeric
#min.thr: numeric, absolute minimal thr, if not provided, by default set to minimal number in "x"
#max.thr: numeric, absolute maximal thr, if not provided, by default set to maximal number in "x"
#returns a vector of logical, same length of input "x". 
#only 1 if that number is within std.thr standard deviation and within absolute range min.thr and max.thr
#otherwise set to 0 
mask_outliers = function(x, std.thr, min.thr = NULL, max.thr = NULL){
    if(is.null(min.thr)){min.thr = min(x)}
    if(is.null(max.thr)){max.thr = max(x)}
    mu = mean(x)
    sigmas = sd(x)
    min.std.thr = mu - std.thr * sigmas 
    max.std.thr = mu + std.thr * sigmas 

    mask = (x <= max.thr) &  (x >= min.thr) & (x <= max.std.thr) & (x >= min.std.thr) 
    return(mask)
}
```

```r
#X: matrix of features by samples, methylation beta values
#y: vector of length number of samples, phenotype of interest to test, 
#W: matrix of number of samples by number of celltypes. 
#cov.mod: matrix of number of samples by number of "global" covariates, to be added to regression 
#max_stds: numberic, samples fall out of the max_stds will be discarded
#this functions essentially fit x ~ 0 + W + yW + cov.mod

#returns a list of three keys
#first key: marg.res, a list of number of celltypes of matrix, each matrix is number of features by coef, SE, t, and p (from T test)
#second key: joint.res, a matrix of number of features by 1. joint p value from a F test
#third key: masks, a matrix of number of features by number of samples. if 1 indicates that sample participate in the regression
celldmc_w_joint = function(X, y, W, cov.mod = NULL, max_stds = 2){
    
    # check pheno
    if (nlevels(factor(y)) == 2) {
        message("Binary phenotype detected. Predicted change will be 1 - 0.")
        y <- factor(y)
    }
    if (!is.factor(y) & !is.character(y)) 
        message("y is not factor or character. Treating as continuous variables.")

    # design matrix
    design <- model.matrix(~ W + y:W)[, -1] # -1 removes the intercept term
    if (!is.null(cov.mod)) design <- cbind(design, cov.mod)
    IntNames.v <- str_c(colnames(W), "Pheno")
    colnames(design)[(1 + ncol(W)):(2*ncol(W))] <- IntNames.v 


    #testing 
    marg.res = array(1,c(nrow(X), ncol(W), 4))
    dimnames(marg.res)[[1]] = rownames(X)
    dimnames(marg.res)[[2]] = IntNames.v
    dimnames(marg.res)[[3]] = c("Estimate", "SE", "t", "p")

    joint.res = matrix(1, nrow(X), 1)
    rownames(joint.res) = rownames(X)
    colnames(joint.res) = c("p")
    
    masks = matrix(0, nrow(X), ncol(X))
    rownames(masks) = rownames(X)
    colnames(masks) = colnames(X)
    
        
    for (feature.id in rownames(X)){
        x.df = as.data.frame(X[feature.id, ])
        colnames(x.df) = c("x")

        # mask outliers
        d1.df = cbind(x.df, design)
        mask  = mask_outliers(d1.df$x, max_stds)
        masks[feature.id, ] = mask
        d1.df = d1.df[mask, ]
        
        # full model
        mdl1.fit = lm(x ~ 0 + ., data = d1.df) 
        marg.res[feature.id,,] = summary(mdl1.fit)$coe[IntNames.v, ]

        # restricted model
        d0.df = d1.df[, !(colnames(d1.df) %in% IntNames.v)]
        mdl0.fit = lm(x ~ 0 + ., data = d0.df) 
        anova.fit <- anova(mdl0.fit, mdl1.fit)
        joint.res[feature.id,] = anova.fit$"Pr(>F)"[2]
    }
    marg.res.list = list()
    for (h in 1:ncol(W)){
        source.id  = colnames(W)[h]
        marg.res.list[[source.id]] = marg.res[,h,]
    }
  
    return(list(marg.res = marg.res.list, joint.res = joint.res, masks = masks))
}
```

# CellDMC has a bug omitting the first column of cov.mod
## gives the same result as if no covars provided 

```r
celldmc.res     <- CellDMC(X, y, W, cov.mod = NULL)
celldmc.res.bug <- CellDMC(X, y, W, cov.mod = cov.mod[,c(1), drop = FALSE])
assert(abs(celldmc.res$coe$Gran[,1:4] - celldmc.res.bug$coe$Gran[,1:4]) < 10**(-10))
assert(abs(celldmc.res$coe$B[,1:4] - celldmc.res.bug$coe$B[,1:4]) < 10**(-10))
```
