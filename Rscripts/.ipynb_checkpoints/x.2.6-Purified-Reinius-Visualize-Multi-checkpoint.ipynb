```r
# merge individaul evaluation bar plots across feature sets: random.10k and hvf.10k
# hvf.10k RMedS
# random.10k correlation and RMedS
```

```r
library("ggplot2")
library("ggpubr")
set.seed(2023)
```

```r
figure.dir  = paste0("../Figure/Purified-Reinius/")
print(figure.dir)
```

# Load 

```r
fig.random.list   = readRDS(file.path(figure.dir, paste0("random.10k/Reinius_eval_random.10k_fig.list.rds")))
stats.random.list = readRDS(file.path(figure.dir, paste0("random.10k/Reinius_eval_random.10k_stats.list.rds")))
fig.hvf.list      = readRDS(file.path(figure.dir, paste0("hvf.10k/Reinius_eval_hvf.10k_fig.list.rds")))
stats.hvf.list    = readRDS(file.path(figure.dir, paste0("hvf.10k/Reinius_eval_hvf.10k_stats.list.rds")))
```

```r
names(fig.random.list)
```

```r
title.size = 20
lab.size = 20
legend.size = 20  
```

```r
options(repr.plot.width = 16, repr.plot.height = 6, repr.plot.res = 200)

hvf.RMedS.bar.g = egg::ggarrange(plots = list(fig.hvf.list$RMedS.bar.plots[["low entropy"]]  + theme(legend.position = "none"), 
                                               fig.hvf.list$RMedS.bar.plots[["high entropy"]]),
          
                   labels = c("a", "b"), 
                   align = "h",
                   widths = c(1,1),  
                   label.args = list(gp = grid::gpar(font = 10, cex =2.5)), 
                   debug=F)


random.cor.bar.g = egg::ggarrange(plots = list(fig.random.list$cor.bar.plots[["low entropy"]]  + theme(legend.position = "none"), 
                                               fig.random.list$cor.bar.plots[["high entropy"]]),
          
                   labels = c("c", "d"), 
                   align = "h",
                   widths = c(1,1),  
                   label.args = list(gp = grid::gpar(font = 10, cex =2.5)), 
                   debug=F)


random.RMedS.bar.g = egg::ggarrange(plots = list(fig.random.list$RMedS.bar.plots[["low entropy"]]  + theme(legend.position = "none"), 
                                                 fig.random.list$RMedS.bar.plots[["high entropy"]]),
          
                   labels = c("e", "f"), 
                   align = "h",
                   widths = c(1,1),  
                   label.args = list(gp = grid::gpar(font = 10, cex =2.5)), 
                   debug=F)
```

# Multi

```r
options(repr.plot.width = 18, repr.plot.height = 16, repr.plot.res = 200)
multi.g = ggarrange(hvf.cor.bar.g, random.RMedS.bar.g, random.cor.bar.g,
                    nrow = 3,
                    heights = c(1,1,1),
                    font.label = list(size = 25, color = "black", face = "bold", family = NULL))
multi.g

ggsave(file.path(figure.dir, paste0("Reinius_eval_multi_supp.pdf")), 
       multi.g,
       device = "pdf", width = 18, height = 16, dpi = 600)
```
